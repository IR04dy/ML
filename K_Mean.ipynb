```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/BankNote_Authentication.csv')

v1 = data['variance']
v2 = data['skewness']
```

```python
V1_V2= np.column_stack((v1,v1))
km_res = KMeans(n_clusters = 2, n_init = 100).fit(V1_V2)
```

```python
clusters = km_res.cluster_centers_
y_kmean = km_res.predict(V1_V2)
```

```python
plt.scatter(v1,v2, c=y_kmean, cmap='viridis')
plt.scatter(clusters[:,0], clusters[:,1], s= 500)
plt.xlabel("V1")
plt.ylabel("V2")

data['kmean'] = km_res.labels_
data['kmean'].value_counts()
```

```python
new_data = data.assign(Class=y_kmean)
print(new_data)
```
